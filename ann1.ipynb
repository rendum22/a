```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#sigmoid activation function 
def sig(x): 
    return 1/(1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 50) 
p = sig(x) 
plt.plot(x, p) 
plt.title("Activation Fucntion: Sigmoid") 
plt.show() 
```
